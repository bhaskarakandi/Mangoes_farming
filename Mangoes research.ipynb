```python
import pandas as pd
import numpy as np

def resample_and_append(df: pd.DataFrame, additional_samples: int) -> pd.DataFrame:
    """
    Resample each column based on its empirical distribution and append the new rows.
    
    Parameters:
    - df (pd.DataFrame): Original DataFrame with n rows and m columns.
    - additional_samples (int): Number of additional rows to generate.
    
    Returns:
    - pd.DataFrame: DataFrame with original and appended resampled rows.
    """
    random_seed = 42
    resampled_data = {}
    np.random.seed(random_seed)
    for col in df.columns:
        # Sample with replacement from the original column
        resampled_data[col] = np.random.choice(df[col].dropna().values, size=additional_samples, replace=True)
    
    # Create a new DataFrame from resampled data
    df_resampled = pd.DataFrame(resampled_data, columns=df.columns)
    
    # Append to the original DataFrame
    df_extended = pd.concat([df, df_resampled], ignore_index=True)
    
    return df_extended
```

```python
import pandas as pd
import numpy as np

def compare_statistics_and_export(df_original: pd.DataFrame, df_extended: pd.DataFrame):
    """
    Compare mean and standard deviation of each numeric column in original and extended DataFrames,
    and export the comparison to an Excel file.

    Parameters:
    - df_original: Original DataFrame (e.g., 18 rows).
    - df_extended: Extended DataFrame (e.g., 270 rows).
    """
    # Select only numeric columns
    numeric_cols = df_original.select_dtypes(include=[np.number]).columns

    # Create a DataFrame for comparison
    comparison = pd.DataFrame(index=numeric_cols)

    comparison['Original_Mean'] = df_original[numeric_cols].mean()
    comparison['Extended_Mean'] = df_extended[numeric_cols].mean()

    comparison['Original_StdDev'] = df_original[numeric_cols].std()
    comparison['Extended_StdDev'] = df_extended[numeric_cols].std()

    comparison['Mean_Difference'] = comparison['Extended_Mean'] - comparison['Original_Mean']
    comparison['StdDev_Difference'] = comparison['Extended_StdDev'] - comparison['Original_StdDev']

    # Optional: export to Excel
    output_filename = "column_statistics_comparison.xlsx"
    comparison.reset_index(inplace=True)
    comparison.rename(columns={'index': 'Column'}, inplace=True)
    comparison.to_excel(output_filename, index=False)

    print(f"Comparison saved to '{output_filename}'")

```
